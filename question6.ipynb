## Qual filme com maior tempo de duração? Calcule os percentils.

```python
import sqlite3
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import cm
from sklearn.feature_extraction.text import CountVectorizer
import warnings
warnings.filterwarnings("ignore")
sns.set_theme(style = "whitegrid")
conn = sqlite3.connect("imdb.db")
```

```python
consulta =  '''
            SELECT runtime_minutes Runtime
            FROM titles
            WHERE type = 'movie' AND Runtime != 'NaN'
            '''
```

```python
resultado = pd.read_sql_query(consulta,conn)
display(resultado)
```

```python
for i in range(101):
    val = i
    perc = round(np.percentile(resultado['Runtime'].values, val), 2)
    print('{} perncetil da duração runtime é :{}'.format(val,perc))
```

```python
# nova consulta para filme de maior duração via SQL
consulta =  '''
            SELECT runtime_minutes Runtime, primary_title
            FROM titles
            WHERE type = 'movie' AND Runtime != 'NaN'
            ORDER BY Runtime DESC
            LIMIT 1
            '''
```

```python
resultado = pd.read_sql_query(consulta,conn)
resultado
```
